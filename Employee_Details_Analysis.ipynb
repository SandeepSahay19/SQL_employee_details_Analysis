## Question 2) _Write an SQL query to find the employee id whose salary lies in the range of 9000 and 15000_.

### Solution:

#### SELECT EmpId, Salary
#### FROM EmployeeSalary
#### WHERE Salary BETWEEN 9000 AND 15000;

## Question 3) _Write an SQL query to know those employees who live in California and work under the manager with ManagerId – 986._

### Solution:

#### SELECT EmpId, City, ManagerId
#### FROM EmployeeDetails
#### WHERE City='Toronto' AND ManagerId='986';

### _RESULT:_  ***Walter White lives in California and Works under Manager Id- 986***

## Question 4) _Write an SQL query to find out all those employees who work on Projects other than P1._

### Solution: 

### SELECT EmpId
### FROM EmployeeSalary
### WHERE NOT Project= 'P1' ;

### _Result:_  ***Walter White (EmpId = 321) and Kuldeep Rana(EmpId=421) works on projects other than P1.***

## Question 5) _Write an SQL query to find the maximum, minimum, and average salary of the employees._

### Solution:

#### SELECT Max(Salary), 
#### Min(Salary), 
#### AVG(Salary) 
#### FROM EmployeeSalary;

### ***Result :***    Maximum Salary is 12000,Minimum Salary is 8000 and Average Salary is 10000.

## _Question 6) Write an SQL query to find all the EmpIds which are present in either of the tables – ‘EmployeeDetails’ and ‘EmployeeSalary’._

### Solution:

#### SELECT EmpId FROM EmployeeDetails
#### UNION 
#### SELECT EmpId FROM EmployeeSalary;

### *RESULT* : EmpId(321),(421) and (121).

## *Question 7. Fetch all the employees who are not working on any project.*

### Solution:

#### SELECT EmpId 
#### FROM EmployeeSalary 
#### WHERE Project IS NULL;

## Question 8. Write an SQL query to fetch all the Employees who are also managers from the EmployeeDetails table.

### Solution:

### SELECT DISTINCT E.FullName
### FROM EmployeeDetails E
### INNER JOIN EmployeeDetails M
### ON E.EmpID = M.ManagerID;
